```python
from fastapi import FastAPI
from pydantic import BaseModel

# =========================
# Configuração da API
# =========================
app = FastAPI(title="Calculadora API", version="1.0")

# Modelo de entrada
class Operacao(BaseModel):
    numero1: float
    numero2: float
    tipo: str  # soma, subtracao, multiplicacao, divisao

# =========================
# Rota principal
# =========================
@app.post("/calcular")
def calcular(op: Operacao):
    resultado = None
    tipo = op.tipo.lower()

    if tipo == "soma":
        resultado = op.numero1 + op.numero2
    elif tipo == "subtracao":
        resultado = op.numero1 - op.numero2
    elif tipo == "multiplicacao":
        resultado = op.numero1 * op.numero2
    elif tipo == "divisao":
        if op.numero2 == 0:
            return {"erro": "Divisão por zero não é permitida"}
        resultado = op.numero1 / op.numero2
    else:
        return {"erro": "Tipo de operação inválido. Use soma, subtracao, multiplicacao ou divisao"}

    return {
        "numero1": op.numero1,
        "numero2": op.numero2,
        "tipo": tipo,
        "resultado": resultado
    }
```

```python
import json

# Dados que você quer salvar
dados = {
    "numero1": 10.5,
    "numero2": 2.0,
    "tipo": "multiplicacao"
}

# Criando o arquivo JSON
with open("exemplo_operacao.json", "w") as arquivo:
    json.dump(dados, arquivo, indent=4)

print("Arquivo 'exemplo_operacao.json' criado com sucesso!")
```
