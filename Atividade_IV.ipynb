```python
produtos = [
    {"id": 1, "nome": "Notebook", "categoria": "Eletrônicos", "preco": 3500.00},
    {"id": 2, "nome": "Mouse", "categoria": "Eletrônicos", "preco": 80.00},
    {"id": 3, "nome": "Teclado", "categoria": "Eletrônicos", "preco": 150.00},
    {"id": 4, "nome": "Cadeira", "categoria": "Móveis", "preco": 600.00},
    {"id": 5, "nome": "Mesa", "categoria": "Móveis", "preco": 900.00}
]
```

```python
filtrados = [p for p in produtos if p["preco"] > 500]

for p in filtrados:
    print(p)
```

```python
for p in produtos:
    if p["nome"] == "Notebook":
        p["preco"] = 3200.00
```

```python
produtos = [p for p in produtos if p["nome"] != "Mouse"]
```

```python
from collections import defaultdict

grupos = defaultdict(list)

# agrupar
for p in produtos:
    grupos[p["categoria"]].append(p)

# mostrar resultados
for categoria, itens in grupos.items():
    quantidade = len(itens)
    media = sum(p["preco"] for p in itens) / quantidade

    print(f"Categoria: {categoria}")
    print(f"Quantidade: {quantidade}")
    print(f"Média de preço: {media:.2f}")
    print("-" * 30)
```
